<div class="alert alert-block alert-warning">
Nous allons dans la suite étudier la répartition des notes et le type de lieu. Pour cela, nous voulons distinguer les types suivants : 

* 1 : uniquement café 
* 2 : uniquement restaurant 
* 3 :  uniquement bar 
* 4 : bar et café
* 5 : bar restaurant 
* 6 : café et restaurant
* 7 : café, bar, restaurant

Nous allons pour cela créer d'abord un nouveau tableau à partir du précédent qui ne contiendra que des lieux de type café, bar, restaurant et qui sera muni d'une nouvelle colonne `type` qui indiquera par un entier le type de ce lieu.
</div>

<div class = "alert alert-block alert-info">
Q1.1

* Créer un tableau  `tab_cbr` à partir de `tableau_poi` contenant uniquement les lieux de type restaurant, bar et café et qui ont un prix supérieur à 0 (n'oubliez pas de copier le tableau grâce à `.copy()`). 
    
* Créer une colonne `type` qui ne contient que des 0.
   
* Affecter un 1 à la colonne `type` si le lieu est uniquement un café, 0 sinon.
    
* Affecter un 2 à la colonne `type` lorsque le lieu est uniquement un restaurant. 
    
* Continuer à compléter la colonne type en mettant un 3 pour les lieux uniquement de type bar, 4 pour bar et café en poursuivant avec toutes les combinaisons possibles.
 
</div>

```python
tab_cbr = tableau_poi[["cafe","restaurant","bar","prix","note","latitude","longitude"]][tableau_poi["prix"]>0].copy()
tab_cbr = tab_cbr.reset_index()

#######
tab_cbr["type"] = np.zeros(len(tab_cbr))

#######
for i in range(len(tab_cbr)):
    if tab_cbr["cafe"][i] == 1 and tab_cbr["restaurant"][i] == 0 and tab_cbr["bar"][i] == 0:
        tab_cbr["type"][i] = 1
    elif tab_cbr["cafe"][i] == 0 and tab_cbr["restaurant"][i] == 1 and tab_cbr["bar"][i] == 0:
        tab_cbr["type"][i] = 2
    elif tab_cbr["cafe"][i] == 0 and tab_cbr["restaurant"][i] == 0 and tab_cbr["bar"][i] == 1:
        tab_cbr["type"][i] = 3
    elif tab_cbr["cafe"][i] == 1 and tab_cbr["restaurant"][i] == 0 and tab_cbr["bar"][i] == 1:
        tab_cbr["type"][i] = 4
    elif tab_cbr["cafe"][i] == 1 and tab_cbr["restaurant"][i] == 1 and tab_cbr["bar"][i] == 0:
        tab_cbr["type"][i] = 5
    elif tab_cbr["cafe"][i] == 0 and tab_cbr["restaurant"][i] == 1 and tab_cbr["bar"][i] == 1:
        tab_cbr["type"][i] = 6
    else:
        tab_cbr["type"][i] = 7
```

```python
cont_cbr = pd.crosstab(tab_cbr["type"],tab_cbr["prix"])
```

<div class = "alert alert-block alert-info">
Q 1.4 - La note n'est pas une variable qualitative - contrairement au prix. Elle correspond à la moyenne des notes données par les utilisateurs (et -1 lorsqu'il n'y aucune note donnée). Afin d'en faire une variable qualitative, créez une nouvelle colonne <code>note_quali</code> dans le tableau <code>tab_bcr</code> qui correspond à la note arrondie par pallier de 0.5 ([1,1.5,2,2.5,...]).
</div>

```python
tab_cbr["note_quali"] = tab_cbr["note"]
lr = [1,1.5,2,2.5,3,3.5,4,4.5,5]

li = []
for i in range(len(tab_cbr)):
    tmp = np.around(tab_cbr["note"][i],0)
    li.append(tmp)

for j in range(len(tab_cbr)):
    for m in range(1,len(lr)):
        if lr[m] == li[j]:
            if tab_cbr["note"][j] <= (lr[m]+lr[m-1])/2:
                tab_cbr["note_quali"][j] = lr[m-1]
            else:
                tab_cbr["note_quali"][j] = lr[m]
```

<div class = "alert alert-block alert-info">
Q 1.5 : Donner le tableau de contingence  entre la note ainsi calculée et le type de lieu
</div>

```python
pd.crosstab(tab_cbr["type"],tab_cbr["note_quali"])
```

<div class = "alert alert-block alert-info">
Q 1.5 - Tracer géographiquement les bars uniquement avec une couleur différente en fonction de la note. Faire de même avec  les restaurants uniquement puis aves les bars/restaurants. Observez vous des différences dans les répartitions des notes en fonction des endroits ?
</div>

```python
def draw(index):
    long = tab_cbr[tab_cbr["type"] == index]["longitude"]
    lat = tab_cbr[tab_cbr["type"] == index]["latitude"]
    l = np.unique(tab_cbr["note_quali"])
    
    
    for i in range(1,6):
        if i == 5:
            plt.scatter(
            tab_cbr[tab_cbr["type"] == index]["longitude"][tab_cbr["note_quali"] == i],
            tab_cbr[tab_cbr["type"] == index]["latitude"][tab_cbr["note_quali"] == i],
            color=plt.cm.RdYlBu(i*50),s=5.5,label=str(i))
            break
        plt.scatter(
            tab_cbr[tab_cbr["type"] == index]["longitude"][tab_cbr["note_quali"] == i],
            tab_cbr[tab_cbr["type"] == index]["latitude"][tab_cbr["note_quali"] == i],
            color=plt.cm.RdYlBu(i*50),s=5.5,label=str(i))
        plt.scatter(
            tab_cbr[tab_cbr["type"] == index]["longitude"][tab_cbr["note_quali"] == i+.5],
            tab_cbr[tab_cbr["type"] == index]["latitude"][tab_cbr["note_quali"] == i+.5],
            color=plt.cm.RdYlBu((i+.5)*50),s=5.5,label=str(i+.5))
            
            
    plt.scatter(
            tab_cbr[tab_cbr["type"] == index]["longitude"][tab_cbr["note_quali"] == -1],
            tab_cbr[tab_cbr["type"] == index]["latitude"][tab_cbr["note_quali"] == -1],
            color=plt.cm.RdYlBu((5.5)*50),s=5.5,label=str(-1))
    plt.legend()
```

```python
plt.title("Répartition spatial des bars uniquement en fonction des notes")
draw(3)
```

```python
plt.title("Répartition spatial des restaurants uniquement en fonction des notes")
draw(2)
```
